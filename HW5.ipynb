```python
#Q8: 

# Introduction to the Problem

# Relation: 
# This experiment is similar to Fisher's Tea experiment, where a participant claimed to be able to distinguish 
# whether milk or tea was poured first into a cup. In this case, 80 students are asked to distinguish the pouring order of 
# tea and milk.

# Null Hypothesis:
# The null hypothesis assumes that the students are guessing randomly, and thus, the proportion of correct responses is 50%
# H0: p = 0.5

# Alternative Hypothesis:
# The alternative hypothesis posits that the true proportion of correct responses is greater than 50%, suggesting that the 
# students are not guessing and may be able to distinguish between the pouring orders.
# HA: p > 0.5

# Quantitative Analysis

# Observed Test Statistic:
# p̂ = 49/80 = 0.6125

# Standard Error under H0:
# SE ≈ 0.056

# Z-Score:
# z ≈ 2.009

# P-Value:
# p ≈ 0.022

# Conclusion
# The null hypothesis H0 assumed that the students were guessing, meaning that the true population proportion p of correct 
# responses is 0.5. Our analysis yielded an observed sample proportion of p̂ = 0.6125, with z-value of 2.013. Using this test
# statistic, we calculated a one-tailed p-value of approximately 0.0214. Since the p-value of 0.022 is less than the typical 
# significance level, 𝛼=0.05, we reject the null hypothesis and accept the alternative hypothesis. This suggests that there 
# is evidence that more than 50% of the students can correctly identify the pouring order.
```

```python
from scipy import stats

# Calculate the p-value for the z-score
p_value = (1 - stats.norm.cdf(abs(2.013)))

p_value
```
